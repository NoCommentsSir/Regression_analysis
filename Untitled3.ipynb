```python
import pandas as pd
import numpy as np
import matplotlib
import sklearn as sk
```

```python
df = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
print(len(df), len(test))
```

```python
train_df = pd.concat([
    df, test
])
train_df['car_mark'].value_counts()
```

```python
train_df = train_df.drop(['car_id','color'], axis="columns")
train_df
```

```python
import re
```

```python
arr = train_df.engine.to_list()
volume = []
power = []
engine = []
for i in range(len(arr)):
  temp = list(re.split('\u2009/\u2009', str(arr[i])))
  try:
    volume.append(float(re.split(' |\xa0', temp[0])[0]))
  except:
    volume.append(0)
  power.append(int(re.split(' |\xa0', temp[1])[0]))
  engine.append(temp[2])
train_df["engine_volume"] = volume
train_df["engine_power"] = power
train_df["engine_type"] = engine
train_df = train_df.drop(['engine'], axis="columns")
train_df
```

```python
help = train_df.engine_volume.to_list()
c = round(train_df.engine_volume.mean(), 1)
for i in range(len(help)):
  if help[i] == 0:
    help[i] = c
train_df["engine_volume"] = help
```

```python
fix = ['автомат', 'механика', 'вариатор', 'робот']
arr = train_df.transmission.to_list()
for i in range(len(arr)):
  if not(arr[i] in fix):
    arr[i] = "заряд"
train_df['transmission'] = arr
train_df
```

```python
arr = train_df["mileage"].to_list()
for i in range(len(arr)):
  if arr[i] == "Новый":
    arr[i] = 0
  else:
    temp = re.split(' |\xa0', arr[i])
    try:
      arr[i] = int(temp[0]+temp[1])
    except:
      arr[i] = int(temp[0])
train_df["mileage"] = arr
train_df
```

```python
train_df = train_df.drop(["car_mark"], axis='columns')
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
average_by_drive = train_df.groupby(by='drive', as_index=False).mean()
average_by_engine = train_df.groupby(by='engine_type', as_index=False).mean()
average_by_transmission = train_df.groupby(by='transmission', as_index=False).mean()
average_by_body = train_df.groupby(by='body', as_index=False).mean()

plt.figure()
plt.subplot(2, 2, 1)
sns.barplot(average_by_drive, x='drive', y='price')
plt.subplot(2, 2, 2)
sns.barplot(average_by_engine, x='engine_type', y='price')
plt.subplot(2, 2, 3)
sns.barplot(average_by_transmission, x='transmission', y='price')
plt.xticks(rotation=90)
plt.subplot(2, 2, 4)
sns.barplot(x="body", y="engine_power", data=train_df);
plt.xticks(rotation=90)
```

```python
drive_dum = pd.get_dummies(train_df['drive'], drop_first=True)
transmission_dum = pd.get_dummies(train_df['transmission'], drop_first=True)
#body_dum = pd.get_dummies(train_df['body'], drop_first=True)
type_dum = pd.get_dummies(train_df['engine_type'], drop_first=True)
train_df = pd.concat([train_df, transmission_dum, drive_dum, type_dum], axis=1)
train_df.drop(['transmission', 'drive', 'engine_type'], inplace = True, axis=1)
train_df
```

```python
import catboost
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from catboost.core import CatBoostRegressor
from sklearn.svm import SVR
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
Y = df["price"]
X = df.drop("price", axis='columns')
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=1)
```

```python
model1 = RandomForestRegressor()
model3 = CatBoostRegressor()
```

```python
model1.fit(x_train, y_train)
pred1 = model1.predict(x_test)
print(metrics.r2_score(y_test, pred1))
print(metrics.mean_absolute_percentage_error(y_test, pred1.astype('int')))
```

```python
model3 = CatBoostRegressor()
model3.fit(x_train, y_train)
pred3 = model3.predict(x_test)
print(metrics.r2_score(y_test, pred3))
print(metrics.mean_absolute_percentage_error(y_test, pred3.astype('int')))
```

```python
params = {
    'n_estimators': [10, 50, 100],
    'criterion': ['squared_error', 'friedman_mse'],
    'max_depth': [10, 25, 50],
    'bootstrap': [True, False],
    'random_state': [True, False],
}
gs = GridSearchCV(model1, params, cv=5, verbose=3)
ans = gs.fit(x_train, y_train)
ans.best_params_
```

```python
pred = gs.best_estimator_.predict(x_test)
print(metrics.r2_score(y_test, pred))
print(metrics.mean_absolute_percentage_error(y_test, pred.astype('int')))
```

```python
model = CatBoostRegressor()
param = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10],
        'l2_leaf_reg': [5, 7, 9]}
gs = GridSearchCV(model, param, cv=10, verbose=3)
ans = gs.fit(x_train, y_train)
ans.best_params_
```

```python
pred = gs.best_estimator_.predict(x_test)
print(metrics.r2_score(y_test, pred))
print(metrics.mean_absolute_percentage_error(y_test, pred.astype('int')))
```
